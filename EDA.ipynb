```python
import pandas as pd
```

```python
books = pd.read_csv(r"data\03_Library Systembook.csv")
customers = pd.read_csv(r"data\03_Library SystemCustomers.csv")
```

```python
customers.head()
```

```python
def fix_invalid_date(x):
    if x == "32/05/2023":
        return "31/05/2023"
    return x
books["Book checkout"] = books["Book checkout"].apply(fix_invalid_date)
books["Book checkout"] = books["Book checkout"].str.replace('"', '', regex=False)
books.head(20)
```

```python
books["Days allowed to borrow"] = books["Days allowed to borrow"].map({
    "2 weeks": 14
})
books.head(20)
```

```python

date_cols = ["Book checkout", "Book Returned"]
books[date_cols] = books[date_cols].apply(pd.to_datetime, dayfirst=True)
books.head()
```

```python
books = books.dropna(subset="Books")
books.head()
```

```python
books["Days on Loan"] = (books["Book Returned"] - books["Book checkout"]).dt.days
books.head()
```
